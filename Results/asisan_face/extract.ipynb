```python
import re
from collections import defaultdict

# Initialize a dictionary to store arrays for each loss variable
losses = defaultdict(list)

# Regular expression to match the loss variable names and their values
pattern = re.compile(r"(\bl_g_[a-zA-Z0-9_]+|\bl_p_512+|\bl_p_256+|\bl_p_128):\s*([\deE\+\-\.]+)")

# Initialize a variable to manually track the iteration count
manual_iter = 0

# Open the train.txt log file
with open("/home/kusan/Recovering-and-Enhancing-Digital-Facial-Images-/GFPGAN/experiments/train_GFPGANv1_512_simple_archived_20240810_150102/train.txt", "r") as file:
    # Process each line in the file
    for line in file:
        # Find the 'iter' value in the line
        if 'iter:' in line:
            # Match using regular expression
            iter_match = re.search(r"iter:\s*([0-9]+)", line)
            if iter_match:
                current_iter = int(iter_match.group(1))
                # If the current iteration is less than the last recorded one, increment manually
                if current_iter <= manual_iter:
                    manual_iter += 100
                else:
                    manual_iter = current_iter
                losses['iter'].append(manual_iter)
        
        # Find all matches of the loss variable pattern in the line
        matches = pattern.findall(line)
        # Store each match in the respective loss array
        for match in matches:
            var_name = match[0]
            value = float(match[1])
            losses[var_name].append(value)

# Convert the defaultdict to a regular dictionary
losses = dict(losses)

# Example: Print the arrays for each loss variable
for var_name, values in losses.items():
    print(f"{var_name}: {values}")

import json
with open("/home/kusan/Recovering-and-Enhancing-Digital-Facial-Images-/GFPGAN/experiments/train_GFPGANv1_512_simple_archived_20240810_150102/variable1.json", "w") as json_file:
    json.dump(losses, json_file, indent=4)
```
